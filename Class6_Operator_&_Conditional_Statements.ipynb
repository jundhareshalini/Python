```python
#traffic light - red/green/yellow
#police - Yes/no

#police yes, signal is red stop
#police no, signal is red go
#police yes, signal is yellow, slow down
#police no, signal is yellow go
#signal green, go

signal = input("enter light:")
police = input("enter police yes/no:")

if signal == "red":
  if police =='yes':
    print("Stop")
  elif police == "no":
    print("go")
elif signal == "yellow":
  if police =='yes':
    print("slow down")
  elif police == "no":
    print("go")
elif signal == "green":
    print("go")
else:
  print("Invalid signal")
```

# CONTROL STATEMENTS

Break, Continue, Pass

```python
#break

for i in range(5):
  if i == 3:
    break
  print(i)
```

```python
#continue

for i in range(5):
  if i == 3:
    continue
  print(i)
```

```python
#pass

for i in range(5):
  if i == 3:
    pass
  print(i)
```
