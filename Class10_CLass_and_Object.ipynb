```python
class calc():

  def add(self,a,b): #a,b is arguments.... self is not argument
    return a + b

  def sub(self,a,b):
    return a - b

c = calc()  #object
d = c.add(5,4)
print(d)
```

```python
c.sub(5,4)
```

```python
class hotel():

  def floor(self,a):
    if a == 1:
      print("veg")
    if a == 2:
      print("non veg")

  def dish(self,a):
    if a == 1:
      print("dal rice")
    if a == 2:
      print("chicken rice")
```

```python
h = hotel()
h.floor(1)
h.dish(1)
```

```python
h = hotel()
h.floor(1)
h.dish(2)
```

```python
#constructor

class hotel():

  def __init__(self,a):
    print("inside constructor")
    self.a = a

  def floor(self):
    if a == 1:
      print("veg")
    if a == 2:
      print("non veg")

  def dish(self):
    if a == 1:
      print("dal rice")
    if a == 2:
      print("chicken rice")
```

```python
obj = hotel(1)
```

```python
obj.a
```

```python
obj.floor()
obj.dish()
```

```python
obj2 = hotel(2)
obj2.a
```

```python
#create construcot
#addition of two number

class cal():

  def __init__(self,a,b):
    self.a = a
    self.b = b

  def add(self):
    c = self.a + self.b
    return c
```

```python
obj3 = cal(5,4)

obj3.add()
```
