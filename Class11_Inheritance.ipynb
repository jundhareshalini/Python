```python
#single inheritance
#one parent - one child

class calc(): #parent class
  def add(self,a,b):
    print(a + b)

class calc1(calc): #child class
  def sub(self,a,b):
    print(a - b)

a = calc1()
a.add(5,4)
a.sub(5,4)
```

```python
#multi level inheritance
#grandparent - parent -  child

class calc(): #grandparent class
  def add(self,a,b):
    print(a+b)

class calc1(calc): #parent class
  def sub(self,a,b):
    print(a-b)

class calc2(calc1): #child class
  def mul(self,a,b):
    print(a*b)

a = calc2()
a.add(5,4)
a.sub(5,4)
a.mul(5,4)


```

```python
# multiple inheritance
#Multiple parents - One child

class calc(): # Parent class 1
  def add(self,a,b):
    print(a+b)

class calc1(): # Parent class 2
  def sub(self,a,b):
    print(a-b)

class calc2(calc1, calc2):    # Inherits from both classes
  def mul(self,a,b):
    print(a*b)

a = calc2()
a.add(5,4)
a.sub(5,4)
a.mul(5,4)




```

```python
#Hybrid Inheritance
#Combination of inheritance types

class calc():# Parent class
  def add(self,a,b):
    print(a+b)

class calc1(calc):# First child class
  def sub(self,a,b):
    print(a-b)

class calc2(calc):   # Second child class
  def mul(self,a,b):
    print(a*b)

class calc3(calc1, calc2):  # Child class inherits from Calc1 and Calc2
  def div(self,a,b):
    print(a/b)

obj = calc3()

obj.add(10, 5)
obj.sub(10, 5)
obj.mul(10, 5)
obj.div(10, 5)
```

```python
#Constructor of parent class and child class
#When a child defines __init__(), it replaces the parent's constructor.
#Python does not call the parent's __init__() automatically

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): # Child class inherits Parent
 def __init__(self):
  self.child = "child"

  def test(self):
    print("test")

a = child()
print(a.child)

print(a.name)
```

```python
#multilevel inheritance with constructor

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): #Child class inherits Parent
  def __init__(self):
    super().__init__()
    self.child = "child"

  def test(self):
    print("test")

class grandchild(child):
  def __init__(self):
    super().__init__()
    self.grandchild = "grandchild"


a = grandchild()
print(a.name)
print(a.child)
print(a.grandchild)
```

```python
#multiple inheritance with constructor

class parent():
  def __init__(self):
    self.name = "parent"

class child():
  def __init__(self):
    self.child = "child"

  def test(self):
    print("test")

class grandchild(child,parent):
  def __init__(self):
    super().__init__()
    super(child,self).__init__()
    self.grandchild = "grandchild"


a = grandchild()
print(a.name)
print(a.child)
print(a.grandchild)
```

```python
from abc import ABC, abstractmethod

class cooking(ABC):

  @abstractmethod
  def salt(self):
    pass

  @abstractmethod
  def water(self):
    pass
```

```python
# Calc - ADD, SUB, MUL, DIV
#ADD, SUB - abstract method

from abc import ABC, abstractmethod

class Calc(ABC):

  @abstractmethod
  def Add(self):
    pass

  @abstractmethod
  def Sub(self):
    pass

class Cal(Calc):

  def Add(self,a,b):
    return a + b

  def Sub(self,a,b):
    return a + b

  def Mul(self,a,b):
    return a + b

  def Div(self,a,b):
    return a + b

a = Cal()
a.Add(5,4)
a.Sub(5,4)
a.Mul(5,4)
a.Div(5,4)

```

```python
#

# a = ["apple", "apple", "mango"]

#class fruit():
  #constructor
  #count
  #unique
  #duplicate

lst = ["apple", "apple", "mango","lemon","banana"]

class fruit():

  def __init__(self,lst):
    self.lst = lst

  def get_detail(self,mode):

    if mode == "count":
      d = {}

      for i in self.lst:
        if i in d:
          d[i] = d[i] + 1
        else:
          d[i] = 1

      return d

    elif mode == "unique":
      unique = []

      for i in self.lst:
        if i not in unique:
          unique.append(i)

      return unique

    elif mode == "duplicate":
      unique = []
      duplicate = []

      for i in self.lst:
        if i not in unique:
          unique.append(i)
        else:
          duplicate.append(i)

      return duplicate

    else:
      return "Invalid"

a = fruit(lst)
print(a.get_detail("count"))
print(a.get_detail("unique"))
print(a.get_detail("duplicate"))
```

```python
# write inheritance on calc
# a, b in const
#calc1 : add
#calc2 : sub

#class calc1():
  #constructor
  #add

#class calc2():
  #sub

#class calc3():
  #mul

#class calc4():
  #div

class calc1():

  def __init__(self,a,b):
    self.a = a
    self.b = b

  def add(self):
    print(self.a + self.b)

class calc2(calc1):

  def sub(self):
    print(self.a - self.b)

class calc3(calc2):

  def mul(self):
    print(self.a * self.b)

class calc4(calc3):

  def div(self):
    print(self.a / self.b)

obj = calc4(5,4)
obj.add()
obj.sub()
obj.mul()
obj.div()
```
