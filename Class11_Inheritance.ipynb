```python
#single inheritance
#one parent - one child

class calc(): #parent class
  def add(self,a,b):
    print(a + b)

class calc1(calc): #child class
  def sub(self,a,b):
    print(a - b)

a = calc1()
a.add(5,4)
a.sub(5,4)
```

```python
#multi level inheritance
#grandparent - parent -  child

class calc(): #grandparent class
  def add(self,a,b):
    print(a+b)

class calc1(calc): #parent class
  def sub(self,a,b):
    print(a-b)

class calc2(calc1): #child class
  def mul(self,a,b):
    print(a*b)

a = calc2()
a.add(5,4)
a.sub(5,4)
a.mul(5,4)


```

```python
# multiple inheritance
#Multiple parents - One child

class calc(): # Parent class 1
  def add(self,a,b):
    print(a+b)

class calc1(): # Parent class 2
  def sub(self,a,b):
    print(a-b)

class calc2(calc1, calc2):    # Inherits from both classes
  def mul(self,a,b):
    print(a*b)

a = calc2()
a.add(5,4)
a.sub(5,4)
a.mul(5,4)




```

```python
#Hybrid Inheritance
#Combination of inheritance types

class calc():# Parent class
  def add(self,a,b):
    print(a+b)

class calc1(calc):# First child class
  def sub(self,a,b):
    print(a-b)

class calc2(calc):   # Second child class
  def mul(self,a,b):
    print(a*b)

class calc3(calc1, calc2):  # Child class inherits from Calc1 and Calc2
  def div(self,a,b):
    print(a/b)

obj = calc3()

obj.add(10, 5)
obj.sub(10, 5)
obj.mul(10, 5)
obj.div(10, 5)
```

```python
#Constructor of parent class

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): # Child class inherits Parent
  def test(self):
    print("test")

a = child()
print(a.name)
```

```python
#Constructor of parent class and child class
#When a child defines __init__(), it replaces the parent's constructor.
#Python does not call the parent's __init__() automatically

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): # Child class inherits Parent
 def __init__(self):
  self.child = "child"

  def test(self):
    print("test")

a = child()
print(a.child)

print(a.name)
```

```python

#Constructor of parent class and child class
#Call the parent constructor explicitly using super().

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): #Child class inherits Parent

 def __init__(self):
  super().__init__()
  self.child = "child"

  def test(self):
    print("test")

a = child()
print(a.child)
print(a.name)
```

```python
#multilevel inheritance with constructor

class parent():
  def __init__(self):
    self.name = "parent"

class child(parent): #Child class inherits Parent
  def __init__(self):
    super().__init__()
    self.child = "child"

  def test(self):
    print("test")

class grandchild(child):
  def __init__(self):
    super().__init__()
    self.grandchild = "grandchild"


a = grandchild()
print(a.name)
print(a.child)
print(a.grandchild)
```
