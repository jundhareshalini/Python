```python
#indexing
```

```python
a = "SachinTendulkar"
```

```python
print(a[0])
print(a[5])
```

```python
#start:end
print(a[0:5])
print(a[5:10])
```

```python
#start:end:interval

print(a[:])
print(a[::1])
print(a[::2]) #Every 2nd character
print(a[::3]) #Every 3rd character
```

```python
#reverse index
print(a[-1]) #guves last character
```

```python
print(a[::-1])
```

```python
#position of character
a = "Sachin Tendulkar"
print(a.index("a")) #gives 1st occurance index
print(a.index("k"))
print(a.index(" "))
print(a.index("a",2)) #gives 2nd occurance index
```

```python
print(a.rindex("T")) #searches from the end of the string, but the index it returns is based on the string's starting position (0-based indexing)."
print(a.rindex("a"))
```

```python
b = "Virat i Kohli"

print(b.index("i"))
print(b.index("i",2))
print(b.index("i",11))
```

```python
print(b.rindex("i"))
```

```python
#remove space for beginning and ending

str = " Hi Hello "

print(str.strip()) # Removes spaces from both the beginning and the end.
print(str.rstrip()) #Removes spaces from the right.
print(str.lstrip()) #Removes spaces from the left.
```
