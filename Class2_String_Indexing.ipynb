```python
#Check whether a string is a palindrome.

a = "abc"

if a == a[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
#Check whether a string is a palindrome.

text = input("Enter a string: ")

reverse = ""

for ch in text:
  reverse = ch + reverse

if text == reverse:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
#Count the occurrence of each character using dictionary.

text = input("Enter a string: ")

d = {}

for i in text:
  if i in d:
    d[i] = d[i] + 1
  else:
    d[i] = 1

print(d)
```

```python
#Count the occurrence of each character without using dictionary.

text = input("Enter a string: ")

a = ""

for i in text:
  if i not in a:
    count = 0

    for j in text:
      if i == j:
        count += 1

    print(i,":",count)
    a += i
```

```python
#Remove duplicate characters.

text = input("Enter a string: ")

a = ""

for i in text:
  if i in a:
    continue
  else:
    a = a + i

print(a)
```

```python
#Remove duplicate characters.

text = input("Enter a string: ")

a = ""

for i in text:
    if i not in a:
      a += i
print(a)
```

```python
#Find the first non-repeated character.

text = input("Enter a string: ")

for i in text:
  count  = 0
  for j in text:
    if i == j:
      count += 1

  if count == 1:
    print("First non-repeated character:", i)
    break
```

```python
#Find the longest word in a sentence.

text = input("Enter a sentence: ")
word = text.split()
longest = ""

for i in word:
  if len(i) > len(longest):
    longest = i

print("Longest word:", longest)
```

```python
#Check if two strings are anagrams.

str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

if sorted(str1) == sorted(str2):
    print("Anagram")
else:
    print("Not an Anagram")
```

```python
#Check if two strings are anagrams without using sorted().

str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

l = list(str2)

for i in str1:
  if i in l:
    l.remove(i)
  else:
    print("Not anagram")
    break
else:
  print("Anagram")

```

```python
#Replace spaces with underscores without replace function

text = input("Enter a sentence: ")

ch = ""

for i in text:
  if i == " ":
    ch = ch + "_"
  else:
    ch = ch + i

print(ch)
```

```python
#Count uppercase and lowercase letters.

text = input("Enter a string: ")

lower = 0
upper = 0

for i in text:
  if i.isupper():
    upper += 1
  elif i.islower():
    lower += 1

print(lower)
print(upper)
```

```python
#Count uppercase and lowercase letters and digit Without using isupper() and islower()

text = input("Enter a string: ")

lower = 0
upper = 0
digit = 0

for i in text:
  if i >= "A" and i <= "Z":
    upper += 1
  elif i >= "a" and i <= "z":
    lower += 1
  elif i >= "1" and i <= "9":
    digit += 1

print(lower)
print(upper)
print(digit)
```

```python
# Count vowels and consonants

text = (input("Enter a sentence: "))
vovel = 0
consonants = 0

for i in text:
  if i.lower() in "aeiou":
    vovel += 1
  elif i.isalpha():
    consonants += 1

print(vovel)
print(consonants)
```

```python
# Count vowels and consonants without using isalpha

text = (input("Enter a sentence: "))
vovel = 0
consonants = 0

for i in text:
  if i in "AEIOUaeiou":
    vovel += 1
  elif "A" <= i <="Z" or "a" <= i <= "z":
    consonants += 1

print(vovel)
print(consonants)
```

```python
#Reverse the order of words in a sentence.

text = (input("Enter a sentence: "))

words = text.split()

print(" ".join(words[::-1]))
```

```python
#Reverse each word in a sentence while keeping the word order the same.

text = input("Enter a sentence: ")
word = text.split()

for i in word:
  reverse = ""

  for j in i:
    reverse = j + reverse

  print(reverse, end = " ")
```
