```python
a = {"name":"Akshay","Age":"21","location":"Mumbai"}
```

```python
# adding new key : value

a["demo"] = "test"

print(a)
```

```python
#Updating existing Value

a["name"] = "Shahrukh"
print(a)
```

```python
#delete

del a["demo"]
print(a)
```

```python
#pop

b = a.pop("location")
print(b)
print(a)
```

```python
#popitem - removes last item

a = {"name":"Akshay","Age":"21","location":"Mumbai"}

b = a.popitem()
print(b)
print(a)
```

```python
#update

a = {"name":"Akshay"}
b = {"age":"20"}
a.update(b)   #merge dictionary
print(a)
```

```python
#sort dictionary

d = {"c": 3, "a": 1, "b": 4}

for i in sorted(d):
    print(i, d[i])
```

```python
#sort

d = {"mango": 20, "apple": 10, "orange": 15}

out = {}

for key in sorted(d):
    out[key] = d[key]

print(out)
```
