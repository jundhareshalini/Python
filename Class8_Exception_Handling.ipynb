```python
a = 5
b = 0

print(a/b)
```

```python
#division by zero

a = 5
b = 0

try:
  print(a/b)
except ZeroDivisionError:
  print("denominator is zero")
  b = 1
  print(a/b)
```

```python
a = 5
b = "test"

print(a/b)
```

```python
#division by zero

a = 5
b = "test"

try:
  print(a/b)
except ZeroDivisionError:
  print("denominator is zero")
  b = 1
  print(a/b)
except TypeError:
  print("Type error has error")
```

```python
#handling all errors

a = 5
b = "test"

try:
  print(a/b)
except Exception as e:
  print(e)
```

```python
# raise error

#handling all errors

a = 5
b = "test"

try:
  print(a/b)
except Exception as e:
  raise(e)
```
