```python
a = ("Hi","Hello","Welcome")
```

```python
#type
```

```python
print(type(a))
```

```python
#isinstance

print(isinstance(a,list))
print(isinstance(a,tuple))
```

```python
#len

print(len(a))
```

```python
# min, max

print(min(a))
print(max(a))
```

```python
 # sorted

 print(sorted(a))
```

```python
 # a[0]

 print(a[0])
```

```python
#reverse

print(a[::-1])
```

```python
#Index

a.index("Hi")
```

```python
# Tuples do not support find(), rfind(), or rindex() methods.
# To search for an element in a list, use the index() method.

a.find("Hi")
```

```python
#we cannot insert, delete or change tuple elements

a = (10, 20, 30)

a[0] = 100 #cannot change elements if tuple
```

```python
#COUNT

a = (10, 20, 30,2 , 20)
a.count(20)
```

```python
#Concatenate

a = (1, 2)
b = (3, 4)

c = a + b

print(c)
```

```python
#Tuple Repetition

a = (1, 2)

print(a * 3)
```

```python
##Packing and Unpacking

a = (10, 20, 30)

x, y, z = a

print(x)
print(y)
print(z)
```

```python
#nested tuple

a = ((1, 2), (3, 4))

print(a[0][1])
```

```python
#sum

a = (10, 20, 30)
sum(a)
```

```python
#Convert List to Tuple

a = [10, 20, 30]

b = tuple(a)

print(a)
type(a)
```

```python
#Deleting a Tuple

a = (10, 20, 30)

print(a)

del a
```

```python
#Tuple as Dictionary Key

d = {(1, 2): "Python"}

print(d[(1, 2)])
```
