```python
# We cannot insert value in set using insert()
```

```python
a = {1,2,3,4,5}
b = {3,4,5,6,7}
```

```python
#union, intersection, difference, difference

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(b.difference(a)) #Elements present in first set only.
print(a.symmetric_difference(b)) #Elements present in either set but not both.
```

```python
#isdisjoint

a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b)) #No common elements.
```

```python
#issubset, issuperset

a = {1,2,3}
b = {1,2,3,4,5}

print(a.issubset(b))
print(a.issuperset(b))
```

```python
print(b.issubset(a))
print(b.issuperset(a))
```

```python
#Adding Multiple Elements

a.update([50, 60,20])
print(a)
```

```python
#discard

a.discard(20) # No error if element not found
print(a)
```

```python
a.pop()         # Removes a random element
print(a)
```

```python
#Convert List to Set

a = [1, 2, 2, 3]

b = set(a)
print(b)
```
