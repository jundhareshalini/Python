```python
#a = [["hi","hello","welcome"],[1,2,3],["demo","try"],[4,5,6]]
#out1 = ["hi","hello","welcome","demo","try"]
#out2 = [1,2,3,4,5,6]

a = [["hi","hello","welcome"],[1,2,3],["demo","try"],[4,5,6]]

out1 = []
out2 = []

for i in a:
  string_flag = False
  int_flag = False
  for j in i:
    if isinstance(j,str):
      string_flag = True
    if isinstance(j,int):
      int_flag = True

  if int_flag == True:
    out2.extend(i)
  elif string_flag == True:
    out1.extend(i)

print(out1)
print(out2)
```

```python
#Count frequency of elements

a = ["apple", "apple", "mango", "mango", "orange"]

fruit_dict = {}

for i in a:
  if i in fruit_dict.keys():
    value = fruit_dict.get(i)
    value = value + 1
    fruit_dict[i] = value
  else:
    fruit_dict[i] = 1

print(fruit_dict)
```

```python
# sort list in asc without using function

a = [6,4,2,1,44,67,32,12]

for i in range(len(a)):
  for j in range(len(a)):
    if a[i] < a[j]:
      a[i], a[j] = a[j], a[i]

print(a)
```
