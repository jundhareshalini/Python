```python
# Shallow copy

a = [1,2,3,4]
b = a

a.append(5)
print(a)
print(b)  #reflect all changes in list b
```

```python
#Deep copy

a = [1,2,3,4]
b = a.copy()

a.append(5)
print(a)
print(b)  # changes will not reflect in list b
```

```python
#index

a = ["Hi","Hello","Welcome"]

print(a.index("Hi"))
```

```python
# Lists do not support find(), rfind(), or rindex() methods.
# To search for an element in a list, use the index() method.
```

```python
a = ["Hi","Hello","Welcome"]

print(a.rindex("Hi"))
```

```python
a = ["Hi",1,2]
b = ["hello","welcome"]

a.append(b)   ## Append the entire list 'b' to list 'a'
print(a)
```

```python

a = ["Hi",1,2]
b = ["hello","welcome"]

a.extend(b)  # Merge list 'b' into list 'a'
print(a)
```

```python
#COUNT

a = [1,2,3,4,2,4,2]
print(a.count(2))
```

```python
#Updating Elements

a = [1,2,3,4,2,4,2]

a[0] = 500
print(a)
```

```python
#Packing and Unpacking

a = [10, 20, 30]
x, y, z = a

print(x)
print(y)
print(z)
```
