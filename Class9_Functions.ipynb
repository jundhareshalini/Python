```python
#positional arguments

def test(*args):
  print(args)
  print(args[0])
  print(args[1])
  print(type(args))

test("hi","hello",1,2,3,4)
```

```python
#keyword arguments

def test(**kwargs):
  print(kwargs)
  print(type(kwargs))

test(name = "dinesh", age = "30", test = "abc")
```

```python
#positional arguments and keyword arguments

def test(*args, **kwargs):
  print(args)
  print(kwargs)

test("Hi","hello",1,2,3,4, name = "dinesh", age = "30", test = "abc")
```

```python
def test(start_pos, end_pos, mode):

  l = []

  if mode == "even":
    remainder = 0
  else:
    remainder = 1

  for i in range(start_pos, end_pos):
    if i % 2 == remainder:
      l.append(i)

  return(l)
```

```python
l = test(5,20,"odd")
print(l)
```

```python
l = test(50,100,"even")
print(l)
```

```python
a = ["apple", "apple", "mango", "lemon"]

# o/p

#unique = ["apple", "mango", "lemon"]
#duplicate = ["apple"]
#count = {"apple" : 2, "mango" : 1, "lemon" : 1}

def get_fruit_detail(a, mode):

  d = {}
  unique = []
  duplicate = []

  for i in a:
    if i not in unique:
      unique.append(i)
    else:
      duplicate.append(i)

    if i in d:
      d[i] = d[i] + 1
    else:
      d[i] = 1

  if mode == "unique":
    print(unique)
    return
  elif mode == "duplicate":
    print(duplicate)
    return
  elif mode == "count":
    print(d)
    return

get_fruit_detail(a,"unique")
```

```python
a = "sachintendulkar123"

#o/p
#get_details(a, mode)
#mode = "alpha", "digit"

def get_detail(a, mode):
  string_mode = ""
  number_mode= ""
  for i in a:
    if i.isdigit():
      number_mode += i
    elif i.isalpha():
      string_mode += i

  if mode == "alpha":
    print(string_mode)
    return
  elif mode == "digit":
    print(number_mode)
    return
  else:
    print("please enter either alpha or digit as mode")

get_detail(a,"digit")
```

```python
#list of prime numbers

l = []

for i in range(2,20):
  prime = True
  for j in range(2,i):
    if i % j == 0:
      prime = False

  if prime == True:
    #print(i)
    l.append(i)

print(l)
```
