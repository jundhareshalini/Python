```python
#Encapsulation
#public -  can be used, modified inside and outside the class
#protected - Can be used by parent and child class, it starts with _
#private - can be be used, modified inside the class, it starts with __
```

```python
#public

class test():

  name = "hello"

a = test()
print(a.name)
```

```python
a.name = "demo"
print(a.name)
```

```python
#Private

class test():

  __name = "hello"

a = test()
print(a.name)   #we cannot access private variable outside the class


```

```python
#private

class test():

  __name = "hello"

  def display(self):
    print(self.__name)

  def change(self,a):
    self.__name = a

a = test()

a.display()
```

```python
#changing value of private class variable

a.change("changed value")

a.display()
```

```python
#use of encapsulation

class user_detail():

  __password = "abcd"

  def show(self):
    print(self.__password)

a = user_detail()
a.show()
```

```python
# Private Method

#use of encapsulation

class user_detail():

  __password = "abcd"

  def __show(self):
    print(self.__password)

a = user_detail()
a._user_detail__show()  # access private method
```

```python
#Polymorphism

#Method overloading

class test():

  def first(self,a,b):
    print(a,b)

  def name(self, *args):
    print(args)

a = test()
a.first(1,2)

```

```python
a.first(3,4)
```

```python
a.name(1,2,3,4)
```

```python
#method overriding

class parent():

  def test(self):
    print("parent")

class child(parent):

  def test(self):
    print("child")

a = child()
a.test()   # child only child class value
```
